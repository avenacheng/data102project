#### 1.2.1 Implementation and Results

* The rewards are subgaussian. The UCB algorithm takes in 4 parameters:
    *  *t* - time step, aka number of days
    * *variance* - the variance of all the rewards over time (a list of 10 numbers for the top 10 stations)
    *  *times_pulled* - a list of length 10, initially all 0, getting updated through each iteration. 
    * _rewards_ - a list of 10 lists, each list initially containing 0, then appending new rewards through each iteration. 


* The UCB algorithm is a simulation tool used to choose which location is best for the recruiter to hand out flyers at. It does this by taking the the arm (intersection) with the highest upper confidence bound, which is calculated using the mean of that arm plus the confidence width, which is determined through number of times that arm is pulled (recruiter is sent out), variance of that arm, and the timestep. Once UCB chooses an arm, we receive a reward, and the means and number of times that arm is pulled is updated. We continue this process based on the number of days in our dataset, and the arm with the highest mean ends up being the best arm. Thoughout the process, we also calculate regret as the reward that we would've received by choosing the optimal intersection minus the reward we actually got. Regret is important in helping us determine if there is an optimal arm. If so, then the cumulative regret will look sublinear.


* Means and upperconfidence bounds get updated at each time step. Means are updated for only the arms that are chosen by UCB, and are done by taking the sum of rewards at that time step divided by the number of times that arm as been pulled. Upper confidence can be defined with the following formula:

$$ C_a(T_a(t),\delta)= \hat{\mu}_a(t)+\sqrt{\frac{2\sigma^2}{T_a(t)}\log{\frac{1}{\delta}}}$$

    * delta = 1/t^2
    * mu hat is mean
    * sigma^2 is variance


```python
def UCB_pull_arm(t,variance,times_pulled,rewards):
    """ 
    Implement the choice of arm for the UCB algorithm
    
    Inputs:
    iteration          - iteration of the bandit algorithm
    times_pulled       - a list of length K (where K is the number of arms) of the number of 
                         times each arm has been pulled
    rewards            - a list of K lists. Each of the K lists holds the samples received from pulling each arm up 
                         to iteration t. 
    
    Returns:
    arm                -  integer representing the arm that the UCB algorithm would choose.
    confidence bounds  -  a list of the upper confidence bounds for each arm
    """

    K = len(times_pulled)
    delta = 1/t**2
    
    confidence_bounds = []
    for arm in range(K):
        if times_pulled[arm] == 0:
            confidence_bounds.append(np.inf)
        else:
            confidence_bounds.append(np.mean(rewards[arm]) + np.sqrt(2*variance[arm]/(times_pulled[arm])*np.log(1/delta)))
            
    arm = np.argmax(confidence_bounds)
    means = [np.mean(rewards[i]) for i in range(K)]
    
    return arm , confidence_bounds, means
```

#### 1.2.2 Discussion

* The results can change over different orderings of which arms to pull first. If you are not pulling the best arm first, then it is possible to get a more linear regret when pulling an arm with a mean close to the best arm. This change is more evident in DC than in NY because DC's top two means are closer to each other than NY's top two means.


* I think it is possible to find a better strategy than the one we used. While continuously sending your recruiter to the best location may seem like a good idea, it ignores the fact that other factors, such as type of day (weekend or weekday), weather, and possible traffic can influence the number of people who attend that day. Thus, I think if we were to incorporate those features into our analysis and send our recruiter to the best location given those conditions, we would have an even better strategy.

```python
def most_likely_user(p1, p2, trip, leaked):
    month = trip.month
    year = trip.year
    sex = trip.sex
    start = trip.start
    end = trip.end
    
    leak = leaked[(leaked.year==year) 
                  & (leaked.sex==sex) 
                  & (leaked.month==month)]
    
    probabilities = []
    for i in range(len(leak)):
        z = leak.iloc[i]['zip']
        if z == start:
            p = 1-p1
        else:
            p = p1
        if z == end:
            p = p*(1-p2)
        else:
            p =  p*p2
        probabilities.append(p)
        
    leak['prob'] = probabilities
    leak = leak.sort_values('prob',ascending=False)
    highest_prob = leak.iloc[0]['prob']
    possible_people = leak[leak.prob == highest_prob]
    person = None
    if len(possible_people) > 1:
        person = possible_people.sample(frac=1)['name']
    else:
        person = possible_people.iloc[0]['name']
    return person
```

**My algorithm is based on comparing probabilities that a user took that trip given that trip's details. Using p1 and p2, I first looked at if their trip started and/or ended in the same place as their address. If they did, then their probabilities were 1 minus their respective p. Otherwise, their probabilities were equal to their respective p. I creaked a new column that showed the user and the probability they took that trip, then only kept the users with the max probability. If there was more than one user, I simply took a random sample and returned that user.**

## 2.4 Takeaways

In summary, I found that it is easy to identify what trips users took given as few as three features, as long as those features are grouped together as a unique identifier. Adding an additional feature that may not seem like a good identifier in itself (i.e. zipcode of their home address) can significantly increase our ability to find more identifiable uesrs. When faced with only features of the trip and none of the user, we can still find out the likelihood of which user it was with a certain amount of confidence.


If data is already released, there is nothing we can do about that or the person who holds the data. However, we could add noise to our data or possibly alter/rearrange the data so that it becomes more difficult to identify which users took which rides. For example, instead of zipcodes we could find some way to encrypt it. From a nontechnical standpoint, it would be a good idea to notify users that had their data breached.


In the future, should we decide to release data to the public, it should be in a way such that there are at least x number of people in all possible feature groupings. For example, if we grouped by sex/month/year, then there should be at least 10 people with that unique grouping. Furthermore, we should not include birthdays or personal zipcodes because it can be very easy to identify people through that. If we want to use a unique identifier to indicate the number of trips a general user took, it should not be based on personal attributes, but rather something more vague (e.g. ID number that we generated and has no inherent meaning).
